```python
#@markdown 이미지-영단어 매칭 퀴즈 STRAT!

from PIL import Image
import random
from collections import OrderedDict

# 이미지와 영단어 매칭 퀴즈 데이터
quiz_data = OrderedDict([
    ("introduction","introduction.jpg"),
     ("loyalty", "loyalty.jpg"),
      ("chemical", "chemical.jpg"),
       ("conservation","conservation.jpg"),
        ("corrosion","corrosion.jpg"),
         ("damage","damage.jpg"),
          ("generation","generation.jpg"),
           ("x-ray photography","x-ray photography.jpg"),
            ("cultural heritage","cultural heritage.jpg"),
            ("glue","glue.jpg"),
            ("document","document.jpg"),
            ("pagoda","pagoda.jpg"),
            ("reward","reward.jpg")
])

# 플레이어 점수
score = 0

# 퀴즈 게임 함수
def play_quiz():
    global score

# 퀴즈 게임 함수
def play_quiz():
    # 퀴즈 문제 생성
    word = next(iter(quiz_data))
    image_path = quiz_data[word]

    # 이미지 로드 및 표시
    image = Image.open(image_path)
    image.show()

    # 이미지 크기 조정
    fixed_size = (200, 200)  # 원하는 크기로 수정 가능
    image = image.resize(fixed_size)

    # 사용자 입력 받기
    answer = input("영단어를 입력하세요: ")

    # 정답 확인
    if answer.lower() == word:
       score += 1  # 정답을 맞춘 경우 점수 1 증가
       print("정답입니다!")
    else:
        print("오답입니다! 정답은", word, "입니다.")

    # 다음 문제를 위해 퀴즈 데이터에서 현재 문제 제거
    quiz_data.pop(word)

# 퀴즈 플레이어 실행
while quiz_data:
    play_quiz()

# 플레이어 점수 출력
print("최종점수:", score)
```

```python
#@markdown 이미지-영단어 매칭 퀴즈 STRAT!

from PIL import Image
import random

# 이미지와 영단어 매칭 퀴즈 데이터
quiz_data = {"introduction" :"introduction.jpg",
             "loyalty" : "loyalty.jpg",
             "chemical": "chemical.jpg",
             "conservation":"conservation.jpg",
             "corrosion":"corrosion.jpg",
             "damage":"damage.jpg",
             "generation":"generation.jpg",
             "x-ray":"x-ray.jpg",
             "cultural heritage":"cultural heritage.jpg"
             }

# 퀴즈 게임
def play_quiz():
    # 퀴즈 문제 생성
    word = random.choice(list(quiz_data.keys()))
    image_path = quiz_data[word]

    # 이미지 로드 및 표시
    image = Image.open(image_path)

    # 이미지 크기 조정
    fixed_size = (350, 350)  # 원하는 크기로 수정 가능
    image = image.resize(fixed_size)

    # 이미지 표시
    image.show()
    
    # 사용자 입력 받기
    answer = input("그림을 보고 알맞은 영단어를 입력하세요!: ")

    # 정답 확인
    if answer.lower() == word:
        print("우와! 정답입니다!")
    else:
        print("아이고...! 정답은", word, "입니다!",)

# 퀴즈 플레이어 실행
play_quiz()
```
